```python
 from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
crime_threshold=5000
df1 = pd.read_csv('C:/Users/Asus/NaiveBayes.csv')
df1["SAFE"] = df1["TOTAL IPC CRIMES"].apply(lambda x: 1 if x < crime_threshold else 0)
features = df1.drop(columns=["STATE/UT", "DISTRICT", "YEAR", "TOTAL IPC CRIMES", "SAFE"])

scaler = StandardScaler()
features = scaler.fit_transform(features)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, df1["SAFE"], test_size=0.2, random_state=42)

# Train the Gaussian Naïve Bayes model with hyperparameter tuning
model = GaussianNB()
param_grid = {'var_smoothing': np.logspace(0, -9, num=100)}
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Use the best model to predict safety for each district
best_model = grid_search.best_estimator_
df1["PREDICTED_SAFE"] = best_model.predict(features)

# Calculate accuracy on the test set
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```
